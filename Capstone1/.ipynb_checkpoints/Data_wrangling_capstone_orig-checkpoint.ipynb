```python
# Import all the necessary packages

import pandas as pd
import numpy as np
import time
```

```python
# Read in the csv file containing the training data

trainingSet = pd.read_csv('TrainingSet.csv', header=0) #header=0
trainingSet = trainingSet.drop('Unnamed: 0', axis=1) # drop the first column
trainingSet
```

```python
# Need to clean up the year format

years = list(trainingSet)[:36]
yearClean = [] # Create an empty list to which to add the 4-digit year
for year in years:
    newYear = str(year)[:4]
    yearClean.append(newYear)

years = yearClean # rename the list of cleaned years as the original list for simplicity
#years # list
```

```python
# Replacing messy year notation 

newColumnNames = years
newColumnNames.extend(['Country Name', 'Series Code', 'Series Name'])
```

```python
# Cleaning up column names in original TrainingSet file

trainingSet.columns = newColumnNames
```

```python
# Need to get a list of unique countries

countries = pd.unique(trainingSet['Country Name'])
#countries # array
```

```python
# Need to get a list of unique series codes

seriesCodes = pd.unique(trainingSet['Series Code'])
#seriesCodes # array
```

```python
# Need to get a list of unique series names

seriesNames = pd.unique(trainingSet['Series Name'])
#seriesNames # array
```

```python
# Ensure that the number of codes and the number of names are the same -- every series name has a series code
assert len(seriesCodes) == len(seriesNames)
```
